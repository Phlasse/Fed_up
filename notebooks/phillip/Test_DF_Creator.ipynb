```python
import pandas as pd
import random
```

```python
urls = pd.read_csv("../../Fed_up/data/samples/final_url.csv")
sample_df = pd.read_csv("../../Fed_up/data/samples/recipe_sample_20201118_1137.csv")
```

```python
merged = sample_df.merge(urls, on="recipe_id", how="right").dropna()
```

```python
merged
```

```python
rating = [random.uniform(0.0, 1.0) for i in range(len(merged.recipe_id))]
```

```python
merged["rating"] = rating
```

```python
final_sample_df = merged.sort_values("rating", ascending=False)
```

```python
final_sample_df
```

```python
final_sample_df.to_csv("../../Fed_up/data/samples/my_sample.csv")
```
