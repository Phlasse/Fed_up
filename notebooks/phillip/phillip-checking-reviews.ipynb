# Checking reviews

```python
import pandas as pd
reviews_df = pd.read_csv("data/RAW_interactions.csv")
recipes_df = pd.read_csv("data/RAW_recipes.csv")
```

```python
reviews_df.head(5)
```

```python
recipes_df.rename(columns={"id":"recipe_id"}, inplace=True)
```

```python
recipes_df.head(5)
```

```python
merged_df = recipes_df.merge(reviews_df, on="recipe_id", how="inner")
```

```python
review_merge_df = merged_df.groupby(by="recipe_id")["contributor_id"].apply(lambda x: ','.join(str(x)))
review_merge_df
```
